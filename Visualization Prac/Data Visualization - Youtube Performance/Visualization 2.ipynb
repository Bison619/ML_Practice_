```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('youtube_channel_real_performance_analytics.csv')
```

```python
print(df.isnull().sum())
print(df.duplicated())
print(df.info())
plt.style.use('fivethirtyeight')
```

```python
df.describe()
```

```python
df.head()
```

```python
df['Video Publish Time'] = pd.to_datetime(df['Video Publish Time'])
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x ='Views', y='Subscribers' ,hue='Year', data=df, palette='viridis')
# plt.xticks(ticks=range(0,int(df['Views'].max())+1, 5000))
plt.title('View and subs per year')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(x ='Views', data=df,hue='Day of Week', palette='Set2')
plt.xticks(ticks=range(0,int(df['Views'].max())+1, 100000))
plt.title('Views countplot')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x ='Revenue per 1000 Views (USD)', data=df, showmeans = True)
# plt.xticks(ticks=range(0,int(df['Views'].max())+1, 5000))
plt.title('Revenue per year')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x ='Day of Week',y='Views', data=df)
plt.title('Views per week days')
plt.show()
```

```python
custom_palette = sns.color_palette("Paired", 12)
sns.palplot(custom_palette)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x ='Year',y='Views', data=df, color = 'red')
plt.title('Views per year')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x ='Day of Week',y='Video Duration', data=df)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x ='Watch Time (hours)',y='Views',hue = 'Month', data=df, palette='viridis')
plt.title('Views per year')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(x ='Subscribers',y= 'Average View Duration', data=df)
plt.xticks(ticks=range(0,int(df['Subscribers'].max())+1, 500))
plt.yticks(ticks=range(0,int(df['Average View Duration'].max())+1, 100))
plt.title('Views countplot')
plt.show()
```
