```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("user_behavior_dataset.csv")
```

```python
print(df.info())
print(df.isnull().sum())
print(df.duplicated())
```

```python
df.head()
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['App Usage Time (min/day)'], kde= True, color= 'red', bins= 30)
plt.title("App usage in days")
plt.xlabel('usage')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x= df['Operating System'],palette='viridis')
plt.title("os usage")
plt.show()
```

```python
plt.figure(figsize=(30, 10))
sns.countplot(x='Number of Apps Installed', data=df, palette='viridis')
plt.title('Count of Users by Number of Apps Installed')
plt.xlabel('Number of Apps Installed')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='Age',y='Gender', data=df, palette='winter',showmeans = True)
plt.title('Age and gender')
plt.xlabel('age')
plt.ylabel('gender')
plt.show()
```

```python
plt.figure(figsize=(22, 10))
sns.countplot(x= 'Device Model',hue = 'Gender', palette='summer', data=df)
plt.title('models')
plt.xlabel('models')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(22, 10))
sns.countplot(x= 'Age',hue = 'Gender', palette='summer', data=df)
plt.title('models')
plt.xlabel('models')
plt.ylabel('Count')
plt.show()
```

```python
df.columns = df.columns.str.strip()
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=df,
    x='App Usage Time (min/day)', 
    y='Screen On Time (hours/day)', 
    hue='Gender', 
    palette='viridis', 

)
plt.title('App Usage Time vs. Screen On Time')
plt.xlabel('App Usage Time (min/day)')
plt.ylabel('Screen On Time (hours/day)')
plt.legend(title='User Behavior Class')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x= 'App Usage Time (min/day)',y='Number of Apps Installed' ,hue='Gender',palette='Set2', data = df)
plt.title('app usage vs num app ')
plt.xlabel('app usage')
plt.ylabel('num app')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.boxplot(x= 'Battery Drain (mAh/day)',y='Operating System',palette='Set3', data = df, showmeans = True)
plt.xticks(ticks=range(0,int(df['Battery Drain (mAh/day)'].max())+1, 200))
plt.title('battery drain vs OS ')
plt.xlabel('battery ')
plt.ylabel('Os')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.boxplot(x= 'Data Usage (MB/day)',y='Gender',palette='viridis', data = df, showmeans = True)
plt.xticks(ticks=range(0,int(df['Data Usage (MB/day)'].max())+1, 200))
plt.title('data vs gen ')
plt.xlabel('data ')
plt.ylabel('gen')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.kdeplot(x= 'Data Usage (MB/day)',y= 'Age', cmap='viridis', data = df, fill = True)
plt.xticks(ticks=range(0,int(df['Data Usage (MB/day)'].max())+1, 200))
plt.yticks(ticks=range(0,int(df['Age'].max())+1, 5))
plt.title('data vs age')
plt.xlabel('data ')
plt.ylabel('age')
plt.show()
```

```python
df.columns
```

```python
sns.pairplot(data=df, vars=[ 'App Usage Time (min/day)', 'Screen On Time (hours/day)',
       'Battery Drain (mAh/day)', 'Number of Apps Installed',
       'Data Usage (MB/day)', 'Age'], diag_kind='kde', hue = 'Gender')
```

```python
plt.figure(figsize=(15, 5))
sns.swarmplot(data=df, x='Device Model', y= 'Data Usage (MB/day)', hue='Gender')
plt.title('model data and gender')
plt.xlabel('models')
plt.ylabel('data usage')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df['Device Model'].value_counts().plot.pie(autopct = '%1.1f%%', explode=[0.05,0.05,0.05,0.05,0.05],startangle= 140,colors = ['skyblue','purple'])
plt.title('model')
plt.ylabel('.')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df['Operating System'].value_counts().plot.pie(autopct = '%1.1f%%', explode=[0.03,0.03],startangle= 120,colors = ['skyblue','purple'])
plt.title('Os')
plt.ylabel('.')
plt.show()
```

```python
g = sns.FacetGrid(df, col='Operating System', row='Gender', margin_titles=True, palette='Set2')
g.map(sns.histplot, 'App Usage Time (min/day)', bins=20, kde=False, color='blue')
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('App Usage Time by OS and Gender', fontsize = 10)
plt.show()
```
