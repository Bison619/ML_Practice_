<h2>Picking the 0s and 1s</h2> 

```python
imbalanced = data[data.Class == 0]
y = data[data.Class == 1]
```

<h2>Taking random 492 value to balance out the data</h2> 

```python
balanced = imbalanced.sample(n=492)
```

```python
balanced.shape
```

```python
data2 = pd.concat([balanced, y], axis=0)
```
